This jupyter notebook provides an NLP-based pipeline for analyzing resumes and job descriptions. It extracts:
Skill-related phrases
Experience-related phrases

from two datasets: one containing resumes and another containing job descriptions.

```python
# Load spacy English model
nlp = spacy.load("en_core_web_sm")

# Load the CSV dataset
def load_resume_csv(file_path):
    df = pd.read_csv(file_path)
    return df
```

```python
#Function to extract skills and experience from resume data
def analyze_resume_text(text):
    sentences = sent_tokenize(text) #tokenize 
    skill_phrases = []
    experience_phrases = []

    for sent in sentences:
        doc = nlp(sent)
        
        for chunk in doc.noun_chunks:
            if any(skill_word in chunk.text.lower() for skill_word in ['python', 'machine learning', 'nlp', 'java', 'deep learning']):
                skill_phrases.append(chunk.text)
        
        if 'year' in sent.lower():
            experience_phrases.append(sent)

    return {
        "skills": list(set(skill_phrases)),
        "experience_sentences": list(set(experience_phrases))
    }
```

```python
#Save the resume corpora in text file.
def save_to_corpus(results, filename="resume_corpus.txt"):
    with open(filename, "w", encoding="utf-8") as f:
        for i, res in enumerate(results):
            f.write(f"--- Resume {i + 1} ---\n")
            f.write("Skills:\n")
            f.write(", ".join(res['skills']) + "\n\n")
            f.write("Experience Sentences:\n")
            f.write("\n".join(res['experience_sentences']) + "\n\n")
            f.write("-" * 40 + "\n\n")
```

```python
import pandas as pd
import nltk
import spacy
from nltk.tokenize import sent_tokenize

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Keywords to look for
SKILL_KEYWORDS = ['python', 'java', 'sql', 'machine learning', 'deep learning', 'nlp', 'data analysis']
EXPERIENCE_KEYWORDS = ['year', 'experience', 'experienced', 'proficient']

# Load your dataset
def load_dataset(file_path):
    df = pd.read_csv(file_path, encoding='utf-8', delimiter="\t") 
    return df

# Analyze a single job description
def analyze_job_description(text):
    #tokenization
    sentences = sent_tokenize(text)
    skill_phrases = []
    experience_phrases = []

    for sent in sentences:
        doc = nlp(sent)
        for chunk in doc.noun_chunks:
            if any(skill in chunk.text.lower() for skill in SKILL_KEYWORDS):
                skill_phrases.append(chunk.text.strip())
        
        if any(keyword in sent.lower() for keyword in EXPERIENCE_KEYWORDS):
            experience_phrases.append(sent.strip())

    return {
        "skills": list(set(skill_phrases)),
        "experience_sentences": list(set(experience_phrases))
    }

# Save all extracted info to a text corpus
def save_to_corpus(results, filename="job_description_corpus.txt"):
    with open(filename, "w", encoding="utf-8") as f:
        for i, res in enumerate(results):
            f.write(f"--- Job Description {i + 1} ---\n")
            f.write("Skills:\n")
            f.write(", ".join(res['skills']) + "\n\n")
            f.write("Experience Sentences:\n")
            f.write("\n".join(res['experience_sentences']) + "\n\n")
            f.write("-" * 40 + "\n\n")

# Main function
if __name__ == "__main__":
    file_path = "./job_descriptions.csv"  # Replace this with your real file path (maybe .csv/.tsv)
    df = load_dataset(file_path)
    results = []

    for idx, row in df.iterrows():
        job_text = str(row.get("Job Description", ""))
        if isinstance(job_text, str) and job_text.strip():
            analyzed = analyze_job_description(job_text)
            results.append(analyzed)

    save_to_corpus(results)
    print("✅ Job description corpus saved as 'job_description_corpus.txt'.")
```
